```python
# Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv(r'C:\Users\yousef\Documents\Projects\dataset\AdidasUSSalesDatasets.csv')
```

```python
# Show Data first few rows

df.head()
```

```python
# check missing data

for col in df.columns:
    dt_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(dt_missing*100)))
```

```python
# columns data types

print(df.dtypes)
```

```python
#  Remove duplicate

df.drop_duplicates()
```

```python
# Sort Data By total sales

df.sort_values(by=['Total Sales'], inplace=False, ascending=False)
```

```python
sns.set(font_scale=1.4)
df['Sales Method'].value_counts().plot(kind='bar', figsize=(7, 6), rot=0)
plt.xlabel("Sales Method", labelpad=14)
plt.ylabel("Count of Product", labelpad=14)
plt.title("Count of sold products by Sales method", y=1.02);
```

```python
sns.set(font_scale=1.4)
df['Retailer'].value_counts().plot(kind='line', figsize=(7, 6), rot=0)
plt.xlabel("Retailer", labelpad=14)
plt.ylabel("Count of Product", labelpad=14)
plt.title("Count of sold products by Retailer", y=1.02);
```

```python
# Add Year Column to get 

df['Invoice Date'] = pd.to_datetime(df['Invoice Date'])
df['year'] = df['Invoice Date'].dt.year
```
