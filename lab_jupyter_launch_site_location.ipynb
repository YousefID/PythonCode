Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
# Creating a new column 'marker_color' based on 'class'
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')
spacex_df.head()
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python



# Initialize the marker cluster
marker_cluster = MarkerCluster()

# Loop through each row in the DataFrame and create a marker
for index, record in spacex_df.iterrows():
    # Create a folium.Marker with custom icon color based on the 'marker_color' column
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],  # Coordinates of the launch
        icon=folium.Icon(color='white', icon_color=record['marker_color'])  # Icon color indicates success or failure
    )
    
    # Add the marker to the marker cluster
    marker_cluster.add_child(marker)

# Add marker_cluster to the site map
site_map.add_child(marker_cluster)

# Display the map
site_map


```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
await piplite.install(['geopy'])
from geopy.distance import geodesic

# Example to calculate distance between two points (launch site and another location)
def calculate_distance(lat1, lon1, lat2, lon2):
    return geodesic((lat1, lon1), (lat2, lon2)).kilometers

# Example of how to apply distance calculation for each launch site to a specific point
# Assuming a specific proximity location with latitude and longitude
proximity_location = (28.5623, -80.5774)  # Example coordinates of a location

spacex_df['distance_to_proximity'] = spacex_df.apply(
    lambda row: calculate_distance(row['Lat'], row['Long'], proximity_location[0], proximity_location[1]), axis=1)
spacex_df.head()
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.features import DivIcon
from geopy.distance import geodesic

# Coordinates for the launch site
launch_site_lat, launch_site_lon = 28.5634, -80.5768  # Example launch site coordinates

# Coordinates for the closest coastline (manually identified using MousePosition)
coastline_lat, coastline_lon = 28.56367, -80.57163  # Example closest coastline coordinates

# Calculate the distance between the launch site and the coastline using geopy
distance_coastline = geodesic((launch_site_lat, launch_site_lon), (coastline_lat, coastline_lon)).kilometers

# Create and add a marker at the coastline point with the calculated distance
distance_marker = folium.Marker(
    location=[coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
    )
)

# Initialize the map and add the marker
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=12)
site_map.add_child(distance_marker)

# Display the map with the marker
site_map
```
